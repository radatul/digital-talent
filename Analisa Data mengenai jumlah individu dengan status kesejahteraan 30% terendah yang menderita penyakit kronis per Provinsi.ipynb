### 1. Deskripsi Data

Dataset ini berisi data mengenai jumlah individu dengan status kesejahteraan 30% terendah yang menderita penyakit kronis per Provinsi.

Penjelasan mengenai Variabel pada Dataset ini:
1. kode_provinsi: Kode Provinsi
2. nama_provinsi: Nama Provinsi
3. kelompok_usia: Kelompok Usia (0 - < 15 tahun, 15 - < 45 tahun, 45 - 60 tahun)
4. jenis_kelamin: Jenis Kelamin
5. jumlah_individu: Jumlah Individu yang Menderita Penyakit Kronis pada Kelompok Usia dan Jenis Kelamin tersebut

```python
import pandas as pd
```

```python
data = pd.read_csv('processed-jumlah-penyakit-kronis-30persen-termiskin-indonesia-perjuli2012.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data
```

### 2. Statistik tentang data

```python
data.kode_provinsi.count()
```

```python
data.count()
```

```python
s = data.jumlah_individu.sum()
mi = data.jumlah_individu.min()
ma = data.jumlah_individu.max()

print("Jumlah =",s)
print("Minimal =", mi)
print("Maksimal =", ma)
```

```python
me = data.mean()
md = data.median()

print(me)
```

```python
print(md)
```

### 3. Visualisasi sederhana

```python
ax = data.groupby('kelompok_usia').mean()['jumlah_individu']
ax = ax.plot(kind='bar', title ="Rata-Rata Penduduk miskin Penderita penyakit kronis Berdasarkan Kelompok Usia", figsize=(5,5), legend=True, fontsize=12)
ax.set_xlabel("Usia", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data.groupby('jenis_kelamin').mean()['jumlah_individu']
ax = ax.plot(kind='bar', title ="Rata-Rata Penduduk miskin Penderita penyakit kronis Berdasarkan Jenis Kelamin", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Jenis Kelamin", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
ax = data.groupby('nama_provinsi').mean()['jumlah_individu']
ax = ax.plot(kind='bar', title ="Rata-Rata Penduduk miskin Penderita penyakit kronis Berdasarkan Provinsi", figsize=(10,5), legend=True, fontsize=12)
ax.set_xlabel("Nama Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
