## Jawab

#### 1. Prepare a list of offices sorted by country, state, city.

```python
run("""
    SELECT country, state, city 
    FROM offices
""")
```

#### 2. How many employees are there in the company?

```python
run("""
    SELECT COUNT(*) AS JumlahKaryawan 
    FROM employees
""")
```

#### 3. What is the total of payments received?

```python
run("""
    SELECT SUM(amount) AS TotalBayar 
    FROM payments 
""")
```

#### 4. List the product lines that contain 'Cars'

```python
run("""
SELECT 
    productName, 
    productLine,
    productVendor,
    productDescription,
    quantityInStock,
    buyPrice,
    MSRP
FROM
    products
WHERE
     productLine LIKE '%Cars'
""")
```

#### 5. Report total payments for October 28, 2004.

```python
run("""
SELECT 
    YEAR(paymentDate) AS Tahun,
    MONTH(paymentDate) AS Bulan,
    DAY(paymentDate) AS Hari,
    SUM(amount) AS TotalBayar 
FROM payments 
WHERE YEAR(paymentDate)=2004 AND MONTH(paymentDate)=10 AND DAY(paymentDate)=28
GROUP BY Tahun,Bulan, Hari
ORDER BY TotalBayar
""")
```

#### 6. Report those payments greater than $100,000.

```python
run("""
    SELECT customerNumber, checkNumber, paymentDate,amount 
    FROM payments
    WHERE amount>100000 
    ORDER BY customerNumber, checkNumber, paymentDate,amount 
""")
```

#### 7. List the products in each product line.

```python
run("""
    SELECT productLine, productName, productCode
    FROM products 
    ORDER BY productLine
""")
```

#### 8. How many products in each product line?

```python
run("""
    SELECT productLine,COUNT(*) AS JumlahProduk 
    FROM products 
    GROUP BY productLine
""")
```

#### 9. What is the minimum payment received?

```python
run("""
    SELECT MIN(amount) AS PembayaranTermurah 
    FROM payments 
""")
```

#### 10. List all payments greater than twice the average payment.

```python
run("""
    SELECT * FROM payments
    WHERE amount > ((SELECT avg(amount)*2 FROM payments)) 
""")
```

#### 11. What is the average percentage markup of the MSRP on buyPrice?

```python
run("""
    SELECT
    (SELECT AVG((msrp-BuyPrice)/buyPrice)*(100)) AS RataRata
    FROM products
""")
```

#### 12. How many distinct products does ClassicModels sell?

#### 13. Report the name and city of customers who don't have sales representatives?

```python
run("""
    SELECT customerName,City,salesRepEmployeeNumber 
    FROM Customers 
    WHERE salesRepEmployeeNumber is Null
""")
```

#### 14. What are the names of executives with VP or Manager in their title? Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.

```python
run("""
    SELECT CONCAT(firstName,' ',lastName) AS fullName, jobTitle
    FROM employees
    WHERE jobtitle like '%VP%' OR jobtitle LIKE '%Manager%'
""")
```

#### 15. Which orders have a value greater than $5,000?
